```python
# (상하, 좌우)대칭, 회전, 노이즈, 밝기

import time
import os
from PIL import Image
from PIL import ImageEnhance
import sys
import random
import glob
import numpy as np
import matplotlib.pyplot as plt


# 프로그램 시작 안내
print("Process Start.")
start_time = time.time()

# 현재 위치 확인
%pwd
# 원하는 경로로 이동
%cd F:\mul_image_crawling\jaeyeong

# 결과 저장 폴더 생성
out_dir = input("make dir name : ")
if out_dir not in os.listdir():
    os.mkdir(out_dir)

    
# 부풀릴 이미지 폴더 불러오기
# glob 이용
img_dir = r"F:\mul_image_crawling\jaeyeong\wound_image2_count_910\*.jpg"
img_list = glob.glob(img_dir)
print(len(img_list))

# 저장된 파일 개수 카운터
COUNT = 1

# 이미지 증강 실행
# 상하반전, 좌우반전, 회전 총 3가지 방법 진행
for img_file in img_list:
    img = Image.open(img_file)
    img = img.convert("RGB")
    img = img.copy()
    Xdim, Ydim = img.size

    # 폴더 내 이미지 모두 읽어와 좌우대칭 저장
    # 이미지 좌우 반전
    img_mirror = img.transpose(Image.FLIP_LEFT_RIGHT)
    # 좌우 반전된 이미지를 저장
    img_mirror.save(out_dir + "\\img_mirror" + "%05d.jpg" %COUNT)


    # 리스트 이미지 모두 읽어와 상하대칭 저장
    # 이미지 상하 반전
    image_updown = img.transpose(Image.FLIP_TOP_BOTTOM)
    # 상하 반전 이미지 저장
    image_updown.save(out_dir + "\\img_updown" + "%05d.jpg" %COUNT)

    
    # 리스트 이미지 모두 읽어와 랜덤으로 회전
    # 랜덤회전 각도 조정
    ang = [-90, -45, 45, 90]
    # 랜덤 회적 각도로 사진 회전
    img_ang = img.rotate(int(random.choice(ang)))    
    # 이미지 크기가 변경 방지 위해 resize()를 실행합니다.
    img_ang = img_ang.resize((Xdim, Ydim))
    # 회전 된 이미지 저장
    img_ang.save(out_dir + "\\img_ang" + "%05d.jpg" %COUNT)
    
    
    # 리스트 이미지 모두 읽어와 흑백의 랜덤 노이즈 추가
    # 이미지 복사
    img_noise = img.copy()
    # 원본 이미지의 사이즈값 저장
    w = img_noise.width ; h = img_noise.height
    # 노이즈가 될 점들 개수 지정
    n = int((w*h)/20)
    # 노이즈가 될 점이 들어갈 랜덤 위치 지정
    x, y = np.random.randint(0, img.width, n), np.random.randint(0, img.height,n)
    # 랜덤 위치에 0 또는 255값 넣어줌
    for (x,y) in zip(x,y):
        img_noise.putpixel((x, y), ((0,0,0) if np.random.rand() < 0.5 else (255,255,255))) 
    # 노이즈 들어간 이미지 저장
    img_noise.save(out_dir + "\\img_noise" + "%05d.jpg" %COUNT) 
    
    
    # 리스트 이미지 모두 읽어와 랜덤 밝기 조정
    br = [0.5, 1.5]
    img_bright = ImageEnhance.Brightness(img)
    img_bright.enhance(float(random.choice(br))).save(out_dir + "\\img_bright" + "%05d.jpg" %COUNT)
    
    
    
    # 저장된 파일 개수 카운터 추가
    print("진행 사진 :", COUNT)
    COUNT += 1

    
    
    
    
                

print("Process Done.")

end_time = time.time()
print("Running Time " + str(end_time - start_time) + " seconds.")



```
