```python
import time
import os
from PIL import Image
import sys
import random
import glob

# 프로그램 시작 안내
print("Process Start.")
start_time = time.time()

# 현재 위치 확인
%pwd
# 원하는 경로로 이동
%cd d:

# 결과 저장 폴더 생성
out_dir = input("make dir name : ")
if out_dir not in os.listdir():
    os.mkdir(out_dir)

    
# 부풀릴 이미지 폴더 불러오기
img_dir = r"F:\mul_image_crawling\jaeyeong\wound_image_count_500\*.jpg"
img_list = glob.glob(img_dir)
print(len(img_list))

# 저장된 파일 개수 카운터
COUNT = 1

# 이미지 증강 실행
# 상하반전, 좌우반전, 회전 총 3가지 방법 진행
for img_file in img_list:
    img = Image.open(img_file)
    img = img.convert("RGB")
    img = img.copy()
    Xdim, Ydim = img.size

    # 폴더 내의 이미지 모두 읽어와 좌우대칭 저장
    # 이미지 좌우 반전
    img_mirror = img.transpose(Image.FLIP_LEFT_RIGHT)
    # 좌우 반전된 이미지를 저장
    img_mirror.save(out_dir + "\\img_mirror" + "%05d.jpg" %COUNT)


    # 리스트 안의 이미지 모두 읽어와 상하대칭 저장
    # 이미지 상하 반전
    image_updown = img.transpose(Image.FLIP_TOP_BOTTOM)
    # 상하 반전 이미지 저장
    image_updown.save(out_dir + "\\img_updown" + "%05d.jpg" %COUNT)

    
    # 리스트 안의 이미지 모두 읽어와 랜덤으로 회전
    # 랜덤회전 각도 조정
    ang = [-90, -45, 45, 90]
    # 랜덤 회적 각도로 사진 회전
    img_ang = img.rotate(int(random.choice(ang)))    
    # 이미지 크기가 변경 방지 위해 resize()를 실행합니다.
    img_ang = img_ang.resize((Xdim, Ydim))
    # 회전 된 이미지 저장
    img_ang.save(out_dir + "\\img_ang" + "%05d.jpg" %COUNT)
    print("진행 사진 :", COUNT)
    
    # 저장된 파일 개수 카운터 추가
    COUNT += 1
    
                

print("Process Done.")

end_time = time.time()
print("Running Time " + str(end_time - start_time) + " seconds.")
```
