#####  색상모드 변경

RGB(Color Model)
RGB 색상 모델은 빛의 삼원색인 Red, Green, Blue를 기본으로 사용하는 모델입니다. 컬러 이미지를 표현할 때 가장 많이 사용되며, 각 채널은 0~255의 값을 가집니다.

HSV(Hue, Saturation, Value)
HSV 색상 모델은 색상(Hue), 채도(Saturation), 명도(Value) 세 가지 요소로 색을 표현하는 모델입니다. 이 모델은 색상, 밝기, 채도를 각각 다룰 수 있기 때문에 컬러 이미지 분석에서 효과적으로 사용됩니다.

LAB(Color Space)
LAB 색상 모델은 인간의 시각 체계에 기반한 색상 모델로, 색도(Lightness), 적색-녹색 축(Red-Green Axis), 청색-황색 축(Blue-Yellow Axis)으로 색상을 표현합니다. 이 모델은 색의 인식 및 감지에 대한 우리의 인식에 가장 잘 맞춰진 색상 모델로 알려져 있습니다.

HLS(Hue, Lightness, Saturation)
HLS 색상 모델은 색상(Hue), 밝기(Lightness), 채도(Saturation) 세 가지 요소로 색을 표현합니다. RGB 색상 모델과 유사하지만, 밝기와 채도를 분리하여 다룰 수 있습니다.

YUV(Color Space)
YUV 색상 모델은 RGB 색상 모델과 유사하지만, 밝기(명암) 정보와 색상 정보를 분리하여 처리하는 방식으로 동작합니다. 흔히 영상 처리나 비디오 인코딩에서 사용되며, 색상 정보를 줄여 효과적으로 압축할 수 있는 장점이 있습니다.

Lab(1964)
Lab(1964) 색상 모델은 L*, a*, b* 채널로 색상을 표현합니다. L은 색의 밝기를, a는 빨강과 초록의 정도, b*는 노랑과 파랑의 정도를 나타내는 채널입니다.

LCH(Color Space)
LCH 색상 모델은 Lab 색상 모델에서 색조(Hue)와 채도(Chroma)를 추가한 것입니다. 색조는 색상을 나타내며, 채도는 색의 순도와 강도를 나타냅니다.

Luv(Color Space)
Luv 색상 모델은 L*, u*, v* 채널로 색상을 표현합니다. L는 밝기, u는 빨강-초록색을, v는 노랑-파랑색을 나타냅니다. 이 모델은 CIE 1976 Lab 색상 모델과 비슷하지만, 다른 색공간에 있는 색들 간의 거리 차이를 더 정확하게 측정할 수 있습니다.

XYZ(Color Space)
XYZ 색상 모델은 인간의 시각 체계를 고려하지 않은 색상 모델로, 색상을 X, Y, Z 채널로 표현합니다. 이 모델은 모든 색상을 표현할 수 있기 때문에 다른 색상 모델로의 변환에 유용합니다.

YCbCr(Color Space)
YCbCr 색상 모델은 영상 압축 및 디지털 비디오 분야에서 많이 사용되며, 색상 정보를 밝기 정보와 색상 정보로 분리하여 처리합니다. 이 모델은 Y, Cb, Cr 채널로 색상을 표현합니다.

YIQ(Color Space)
YIQ 색상 모델은 NTSC(미국 국립 표준 위원회) 텔레비전에서 사용되는 색상 모델로, Y(밝기), I(초록-빨강 차이), Q(파랑-노랑 차이) 채널로 색상을 표현합니다.

YDbDr(Color Space)
YDbDr 색상 모델은 YUV 색상 모델과 비슷하지만, 색상 정보를 파랑-반대, 노랑-반대, 밝기-차이 채널로 분리하여 처리합니다.

##### 소비판 밝기별로 구분된 폴더에 옮기기

```python
# 필요한 패키지 불러오기
import cv2
import os
import shutil
import numpy as np

# 이미지 디렉토리 경로 설정
image_dir = "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/aihub/images/REG"  # 원본 이미지 경로 설정
image_save = "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/aihub" # 이미지 저장 경로 설정

# 결과를 저장할 디렉토리 생성
# 밝기별로 low, mid, high로 나눔
output_dir_low = os.path.join(image_save, "R_low")
output_dir_mid = os.path.join(image_save, "R_mid")
output_dir_high = os.path.join(image_save, "R_high")

os.makedirs(output_dir_low, exist_ok=True)
os.makedirs(output_dir_mid, exist_ok=True)
os.makedirs(output_dir_high, exist_ok=True)

# 이미지 파일 목록 얻기
image_files = [f for f in os.listdir(image_dir) if f.endswith((".jpg"))] # jpg로 끝나는 파일들 경로 리스트 생성

# 이미지들 밝기 별로 나눈 후 밝기 별 폴더에 복사
for image_file in image_files:
    
    # 이미지 리스트에서 1장씩 불러오기
    image_path = os.path.join(image_dir, image_file)
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # R 채널 히스토그램 계산
    imgRed = img[:,:,0].flatten() #flatten() 메소드로 밝기 별  1차원 리스트로 만듬
    imgRedPixel = round(sum(imgRed)/len(imgRed),1) #리스트 값들을 합친 값을 길이로 나눠 평균값 

    # 히스토그램을 기반으로 이미지 분류
    if imgRedPixel >=100 and imgRedPixel<150:
        output_path = os.path.join(output_dir_high, image_file)
    
    elif imgRedPixel >= 150:
        output_path = os.path.join(output_dir_mid, image_file)
    
    else :
        output_path = os.path.join(output_dir_low, image_file)
    

    # 이미지를 해당 폴더에 복사
    shutil.copy(image_path, output_path)

print("이미지 분류 및 저장이 완료되었습니다.")
```

```python
# 이미지 순서 랜덤 적용 후 train, test, valid 폴더로 분배

import glob
import shutil
import random

# 밝기별 폴더 경로 변수
EGG640ImgRL = "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/aihub/R_low"
EGG640ImgRM = "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/aihub/R_mid"
EGG640ImgRH = "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/aihub/R_high"

# for문을 위해 밝기별 폴더 경로 변수 넣은 리스트 생성
EGG640ImgRLIst=[EGG640ImgRL,EGG640ImgRM,EGG640ImgRH]

# train, test, valid 폴더 경로 변수
image640Train =  "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/images/train"
image640Test =  "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/images/test"
image640Valid =  "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/images/valid"
label640Train =  "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/labels/train"
label640Test =  "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/labels/test"
label640Valid =  "/home/dlawodud0484/hyun/Dataset/EGG640_dataset/labels/valid"


# train
for l in EGG640ImgRLIst:
    imgList = glob.glob(l+"/*.jpg")
    random.shuffle(imgList) # glob 해온 리스트 순서 무작위 섞은 리스트
    
    trainList = imgList[0:900] # train 900장
    for file in trainList:
        file_split = file.split("/")[-1]
        fname_split = file_split.split(".")[0]

        imageTrainsrc = os.path.join(l, fname_split+".jpg")
        imageTraindst = os.path.join(image640Train, fname_split+".jpg")
        shutil.copyfile(imageTrainsrc, imageTraindst)
        labelTrainsrc = os.path.join(EGG640Labels, fname_split+".txt")
        labelTraindst = os.path.join(label640Train, fname_split+".txt")
        shutil.copyfile(labelTrainsrc, labelTraindst)


# test
    testList=imgList[900:1200] # test 300장
    for file in testList:
        file_split = file.split("/")[-1]
        fname_split = file_split.split(".")[0]

        imageTestsrc = os.path.join(l, fname_split+".jpg")
        imageTestdst = os.path.join(image640Test, fname_split+".jpg")
        shutil.copyfile(imageTestsrc, imageTestdst)
        labelTestsrc = os.path.join(EGG640Labels, fname_split+".txt")
        labelTestdst = os.path.join(label640Test, fname_split+".txt")
        shutil.copyfile(labelTestsrc, labelTestdst)

# valid
    validList=imgList[1200:1500] # valid 300장
    for file in validList:
        file_split = file.split("/")[-1]
        fname_split = file_split.split(".")[0]

        imageValidsrc = os.path.join(l, fname_split+".jpg")
        imageValiddst = os.path.join(image640Valid, fname_split+".jpg")
        shutil.copyfile(imageValidsrc, imageValiddst)
        labelValidsrc = os.path.join(EGG640Labels, fname_split+".txt")
        labelValiddst = os.path.join(label640Valid, fname_split+".txt")
        shutil.copyfile(labelValidsrc, labelValiddst)
```

```python
# 색상모델당 채널별 데이터셋 생성

# 필요한 라이브러리 불러오기
import os
import cv2
import sys
import glob
import numpy as np
import matplotlib.pyplot as plt
import shutil
import os
import yaml

# 원본 데이터셋 경로 변수 설정
EGG640_Src = "/home/dlawodud0484/hyun/Dataset/EGG640_dataset"

# 원본 데이터셋 경로 변수 설정
YOLO_EGG640 = "/home/dlawodud0484/hyun/Dataset/YOLO_EGG/EGG640"

# for문을 위해 색상 모델별 폴더 이름 리스트 생성
EGG640_dirList = ["640_G", "640_RGB_R", "640_RGB_G", "640_RGB_B", "640_HSV_H", "640_HSV_S", "640_HSV_V", "640_LAB_L", "640_LAB_A", "640_LAB_B", "640_HLS_H", "640_HLS_L", "640_HLS_S", "640_YUV_Y", "640_YUV_U", "640_YUV_V", "640_LUV_L", "640_LUV_U", "640_LUV_V", "640_XYZ_X", "640_XYZ_Y", "640_XYZ_Z", "640_YCrCb_Y", "640_YCrCb_Cr", "640_YCrCb_Cb"]

# 원본폴더 데이터셋을 채널별 데이터셋으로 옮기기
for EGG640_dir in EGG640_dirList:
    output_dir = os.path.join(YOLO_EGG640, EGG640_dir)
    os.makedirs(output_dir, exist_ok=True) # exist_ok = True는 디렉토리가 기존에 존재하면 에러발생 없이 넘어가고, 없을 경우에만 생성
    dst_dir = os.path.join(YOLO_EGG640, EGG640_dir)
    shutil.copytree(EGG640_Src, dst_dir, dirs_exist_ok=True) # images, labels 디렉토리 복사
    shutil.copy(EGG640_Src+"/data.yaml", dst_dir) # data.yaml 파일복사
 
    # train, test, valid 이미지 경로 넣은 리스트 생성
    train_list = glob.glob(YOLO_EGG640+"/"+EGG640_dir+"/images/train/*.jpg")
    test_list = glob.glob(YOLO_EGG640+"/"+EGG640_dir+"/images/test/*.jpg")
    val_list = glob.glob(YOLO_EGG640+"/"+EGG640_dir+"/images/valid/*.jpg")


    # train, test, valid 이미지 경로 넣은 txt파일 생성
    with open(YOLO_EGG640 +"/"+ EGG640_dir +"/"+"train.txt", "w") as f:
        f.write("\n".join(train_list) + "\n")

    with open(YOLO_EGG640 +"/"+ EGG640_dir +"/"+"test.txt", "w") as f:
        f.write("\n".join(test_list) + "\n")

    with open(YOLO_EGG640 +"/"+ EGG640_dir +"/"+"val.txt", "w") as f:
        f.write("\n".join(val_list) + "\n")

    # data.yaml 불러오기
    with open(YOLO_EGG640 +"/"+ EGG640_dir+"/"+"data.yaml","r") as ya:
        data = yaml.full_load(ya)

    # data.yaml에 테스트 검증 이미지 주소 넣어주기
    data["train"] = YOLO_EGG640 +"/"+ EGG640_dir +"/"+"train.txt"
    data["test"] = YOLO_EGG640 +"/"+ EGG640_dir +"/"+"test.txt"
    data["val"] = YOLO_EGG640 +"/"+ EGG640_dir +"/"+"val.txt"

    with open(YOLO_EGG640 +"/"+ EGG640_dir+"/"+"data.yaml","w") as ya:
        yaml.dump(data,ya) #dump로 덮어 씌우기
    
    
# 프로그램 시작 안내
print("이미지 변환 시작")

# 저장된 파일 개수 카운터
count = 0

# for문을 위해 train, test, valid 문자열 넣은 리스트 생성
type_dirs = ["train", "test", "valid"]

# 원본 데이터셋 경로, 색상모델 채널별 데이터셋 상위 폴더 경로 변수 설정
original = "/home/dlawodud0484/hyun/Dataset/EGG640_dataset"
dataset = "/home/dlawodud0484/hyun/Dataset/YOLO_EGG/EGG640"


# 색상 모델, 채널별 이름 넣은 리스트 생성
channelTypes = ["RGB", "HSV", "LAB", "HLS" ,"YUV" ,"LUV" ,"XYZ", "YCrCb"]
channelChecks = ["640_RGB_R", "640_RGB_G", "640_RGB_B", "640_HSV_H", "640_HSV_S", "640_HSV_V", "640_LAB_L", "640_LAB_A", "640_LAB_B", "640_HLS_H", "640_HLS_L", "640_HLS_S", "640_YUV_Y", "640_YUV_U", "640_YUV_V", "640_LUV_L", "640_LUV_U", "640_LUV_V", "640_XYZ_X", "640_XYZ_Y", "640_XYZ_Z", "640_YCrCb_Y", "640_YCrCb_Cr", "640_YCrCb_Cb"]

# while문 종료를 위한 숫자 변수 생성
ok1=0
        

# 데이터셋 색상모델 및 채널 별 영상처리 후 저장
while ok1==0: # 0 초과할 경우 정지
    for type_dir in type_dirs:

        for channelType in channelTypes:      
            c = 0 # 채널을 0,1,2 총 3개씩 구분해서 저장하기 위한 숫자 변수 설정 

            for channelCheck in channelChecks:

                if channelType in channelCheck: # channelCheck의 문자열 중에 channelType 문자열이 들어가는지 확인  ex: if "RGB" in "640_RGB_R" : 이 때 if문은 True값을 가짐
                    img_dir = dataset+"/"+channelCheck+"/images/"+type_dir+"/"+"*.jpg"
                    print("현재 경로는", img_dir, "입니다")
                    img_list = glob.glob(img_dir)
                    print("총",len(img_list),"장 변환 시작합니다")

                    # 영상처리 시작
                    for img_file in img_list:
                        img = cv2.imread(img_file)
                        imgChannelType = cv2.cvtColor(img, getattr(cv2, f'COLOR_BGR2{channelType}')) # getattr()함수를 이용해 문자열 리스트 변수를 인수로 사용
                        imgOutput = imgChannelType[:,:,c]
                        cv2.imwrite(img_file, imgOutput) # 이미지 저장 함수
                        count += 1
                    print(channelCheck,"채널 중",c,'채널 저장 했습니다')
                    c = c+1 # 다른 채널로 영상처리를 해주기 위해 1값을 더 해줌

                else:
                    pass

    ok1 +=1
print("총 %d 장 파일을 만들었습니다"%count)
```
